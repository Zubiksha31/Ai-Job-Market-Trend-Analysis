```python
skill_cols = ["skills_python","skills_sql","skills_ml","skills_deep_learning","skills_cloud"]
skill_demand = df[skill_cols].sum().sort_values(ascending=False)

colors = ["#6FC995", "#358DB0", "#358DB0", "#358DB0",  "#DB6B6B"]

fig, ax = plt.subplots(figsize=(10,5)) 

bars = ax.bar(skill_demand.index, skill_demand.values, color=colors[:len(skill_demand)],  width=0.6)

# Labels & title
plt.xlabel("Skills", fontsize=12, fontweight="bold", labelpad=10)
plt.ylabel("Demand Count", fontsize=12, fontweight="bold", labelpad=10)
plt.title("Skill Demand Analysis", fontsize=14, fontweight="bold", pad=15)
plt.xticks(rotation=0, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.4)

max_val = skill_demand.max()
min_val = skill_demand.min()
ax.set_ylim(0, max_val * 1.25)
for bar in bars:
    height = bar.get_height()
    
    # HIGH → Green
    if height == max_val:
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height + (max_val * 0.05),
            f'High: {int(height)}',
            ha='center',
            fontsize=11,
            fontweight='bold',
            color="green"
        )
    elif height == min_val:
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height + (max_val * 0.05),
            f'Low: {int(height)}',
            ha='center',
            fontsize=11,
            fontweight='bold',
            color="red"
        )

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```
